# DP Mechanisms Example (Python)
This notebook demonstrates how to use the `LaplaceNoise` and `GaussianNoise` mechanisms from the Differential Privacy Python library.

```python
# Install the library (if not already installed)
# !pip install differential-privacy

from dp_mechanisms import LaplaceNoise, GaussianNoise

# Laplace example
laplace = LaplaceNoise(epsilon=1.0)
noisy_val = laplace.add_noise(42.0)
print('Laplace noisy value:', noisy_val)

# Gaussian example
gaussian = GaussianNoise(epsilon=1.0, delta=1e-5)
noisy_val_g = gaussian.add_noise(42.0)
print('Gaussian noisy value:', noisy_val_g)
```

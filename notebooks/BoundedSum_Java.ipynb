# BoundedSum Example (Java)
This notebook demonstrates how to use the `BoundedSum` algorithm from the Differential Privacy Java library.

```java
// Java example (run with javac and java)
import com.google.privacy.differentialprivacy.BoundedSum;
public class BoundedSumExample {
  public static void main(String[] args) {
    BoundedSum.Builder builder = BoundedSum.builder();
    builder.setEpsilon(1.0).setLower(0L).setUpper(100L);
    BoundedSum sum = builder.build();
    sum.addEntry(42L);
    long result = sum.computeResult();
    System.out.println("DP sum: " + result);
  }
}
```
